Step Function without numpy

```python
# Defining Step function here

def step(x):
    if x > 0:
        return 1
    else:
        return 0
    
# Using Step Function

step(-200)
```

Step Function using numpy

```python
import numpy as np
# step_function numpy version
def step(x):
    return np.array(x > 0).astype(int)

step_arr = np.array([1, 200, -2, 0, 0.5])

step(step_arr)
```

Plotting Step 

```python
import matplotlib.pyplot as plt

x = np.arange(-10.0, 10.0, 0.1)
y = step(x)
```

```python
plt.plot(x, y)
plt.title('Step Function Plot')
plt.show()
```

Sigmoid Function

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

Plotting Sigmoid Function

```python
y = sigmoid(x)
plt.plot(x, y)
plt.show()
```

Making some functions do Multilayer perceptron easily

```python
def init_network():
    net = {}
    # layer 1
    net['w1'] = np.array([[0.3, 0.33, 0.3],[0.5, 0.4, 0.1]])
    net['b1'] = np.array([1, 1, 1])
    # layer 2
    net['w2'] = np.array([ [0.2, 0.3], [0.4, 0.45], [0.22, 0.1234] ])
    net['b2'] = np.array([0.5, 0.5])
    # layer 3 <-- output
    net['w3'] = np.array([ [0.7, 0.91], [0.123, 0.314] ])
    net['b3'] = np.array([0.1, 0.2])

    return net

def forward(net, x):
    w1, w2, w3 = net['w1'], net['w2'], net['w3']
    b1, b2, b3 = net['b1'], net['b2'], net['b3']

    a1 = np.dot(x, w1) + b1
    z1 = sigmoid(a1)

    a2 = np.dot(z1, w2) + b2
    z2 = sigmoid(a2)

    a3 = np.dot(z2, w3) + b3
    y = identity(a3)

    return y
```

```python
net = init_network()
y = forward(net, np.array([7.0, 2.0]))
y
```
